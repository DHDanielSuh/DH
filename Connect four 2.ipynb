```python
import random
```

```python
class TicTacToe():
    
    def __init__(self):
        
        self.players = ('Player 1', 'Player 2')
        self.board = [[x] for x in range(1, 43)]
      
    
    def display_board(self):
        
        n = 0
        while n < len(self.board):
            print('\t\t', self.board[n], self.board[n+1], self.board[n+2],  self.board[n+3], self.board[n+4], self.board[n+5], self.board[n+6])
            n += 7
        
        print("")
    
    
    def choose_turn(self):
        
        first_move = random.choice(self.players)
        print('\t{} starts first.'.format(first_move))
        self.display_board()
        self.place_marker(first_move)
    
    
    def switch_turn(self, player):
        
        if player == self.players[0]:
            player = self.players[1]
        else:
            player = self.players[0]
            
        self.place_marker(player)
    
    
    def main(self):
        
        print("""
                        Tic Tac Toe
                        Player 1 : 0
                        Player 2 : X
        """)
        self.choose_turn()
    
    
    def place_marker(self, player):
        
        if player == self.players[0]:
            marking = " 0 "
        if player == self.players[1]:
            marking = " X "
            
        INVALID_INPUT = 'Invalid input.'
        
        try:
            position = int(input('{} turn.: '.format(player))) - 1
            self.validity_check(position, marking, player)
        except ValueError: 
            print(INVALID_INPUT)
            self.place_marker(player)
        except IndexError:
            print(INVALID_INPUT)
            self.place_marker(player)
        
        self.switch_turn(player)   
    
    
    def validity_check(self, position, marking, player):
        
        if isinstance(self.board[position], str):
            print("Cannot put it here.")
            self.place_marker(player)
        elif position < 35 and type(self.board[position+7]) is list:
            print("Cannot put it here.")
            self.place_marker(player)
        else:
            self.board[position] = marking
            self.display_board()
            self.win_check(player)
            
    
    def win_check(self, player):
        
        if (self.board[0] == self.board[1] == self.board[2] == self.board[3] or
            self.board[3] == self.board[4] == self.board[5] == or
            self.board[6] == self.board[7] == self.board[8] or
            self.board[0] == self.board[3] == self.board[6] or
            self.board[1] == self.board[4] == self.board[7] or
            self.board[2] == self.board[5] == self.board[8] or
            self.board[0] == self.board[4] == self.board[8] or
            self.board[2] == self.board[4] == self.board[6]):
            print(player + " won!!")
            self.replay()
        else:
            self.full_check(player)

    
    def full_check(self, player):
        
        if all(isinstance(x, str) for x in self.board):
            print("Board is full!")
            self.replay()
        else:
            self.switch_turn(player)
        
    
    def replay(self):
        answer = input("Do you want to play again?(Y/N)")
        if answer.lower == "y":
            self.__init__()
            self.choose_turn()
        elif answer == "y":
            self.__init__()
            self.choose_turn()
        else:
            quit()

            
if __name__ == "__main__":
    game = TicTacToe()
    game.main()
```
